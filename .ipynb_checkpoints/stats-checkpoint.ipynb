```python
heros = {}
with open('7.32e.json') as json_file:
    data = json.load(json_file)
    for key in data:
        heros[key] = Hero(key)
```

```python
strheros = {}
for hero in heros:
    if heros[hero].primaryA == "strength":
        strheros[heros[hero].name] = (heros[hero].str,heros[hero].strg)
agiheros = {}
for hero in heros:
    if heros[hero].primaryA == "agility":
        agiheros[heros[hero].name] = (heros[hero].agi,heros[hero].agig)
intheros = {}
for hero in heros:
    if heros[hero].primaryA == "intelligence":
        intheros[heros[hero].name] = (heros[hero].int,heros[hero].intg)
print(strheros)
print(agiheros)
print(intheros)
```
